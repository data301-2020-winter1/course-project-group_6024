# Task 4: EDA OF DATA
In this task we'll be looking at the general overview of the data by using describe, boxplots, histograms and scatter plots.

I will use the popular pandas and numpy libraries to do the Exploratory Analysis

```python
import pandas as glouis
import numpy as david

df = glouis.read_csv("../../data/raw/Medical_Cost.csv")

print(df.head())
```

## Step 1: We will first describe the data using the describe method

```python
df[["age","bmi","children","charges"]].describe()
```

From the describe, we can see that there are 1338 rows, and the average age of the DataFrame is 39.2 years, the average BMI is 30.66, aveage number of children is 1, and lastly average cost was CAD 13270.

Before we begin analysis, we need to get rid of all empty rows

```python
df.dropna().reset_index()
```

```python
df[["age", "bmi"]].plot.box()
```

- From the boxplot, we can see that majority of the age groups lie between 29 to 50, with no outliers.
- The bmi plot, we can see that the bmi ranges between 29 to 35, with a number of outliers.

```python
df[["charges"]].plot.hist(bins=10, title="Distribution of charges")
```

From the histogram, we can see that majority of the patients paid less than CAD 10,000. with a couple paying CAD 60,000

```python
#df.groupby(["region"]).count().plot(kind='bar')
df["region"].value_counts()[:20].plot(kind="bar")
```
